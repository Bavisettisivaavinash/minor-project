```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive



# Specify the file path in Google Drive
file_path = '/ziyadh.csv'

# Load the dataset into a pandas DataFrame
df = pd.read_csv(file_path)

# Strip any leading/trailing spaces from column names
df.columns = df.columns.str.strip()

# Display the first few rows and column names
print(df.head())
print(df.columns)

# Drop unnecessary columns (e.g., 'sample_id') if present
df = df.drop(columns=['id'], errors='ignore')  # Adjust based on actual column names

# Separate features and labels
X = df.drop('class', axis=1)  # Features
y = df['class']  # Labels

# Check the number of features
print(f"Number of features: {X.shape[1]}")

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Check the number of features after scaling
print(f"Number of features after scaling: {X_scaled.shape[1]}")

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Check the number of features in training and test sets
print(f"Training features shape: {X_train.shape}")
print(f"Test features shape: {X_test.shape}")

# Initialize and train the Random Forest classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Predict on the test set
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Detailed classification report
print(classification_report(y_test, y_pred))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()


```

```python
import joblib
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import pandas as pd

# Load the dataset
file_path = '/ziyadh.csv'
df = pd.read_csv(file_path)

# Strip any leading/trailing spaces from column names
df.columns = df.columns.str.strip()

# Drop unnecessary columns if any
df = df.drop(columns=['id'], errors='ignore')

# Separate features and labels
X = df.drop('class', axis=1)
y = df['class']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Save the scaler
joblib.dump(scaler, '/content/drive/My Drive/Colab Notebooks/scaler.pkl')

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize and train the classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Save the model
joblib.dump(clf, '/content/drive/My Drive/Colab Notebooks/parkinsons_detector.pkl')
```
